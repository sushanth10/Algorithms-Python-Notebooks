# Floyd's Algorithm

This algorithm is used to find the shortest path between any two given vertices of a graph.

```python
import math
def InpWgtMat(n):
    mat = [[0 for i in range(n)] for j in range(n)]
    for i in range(n):
        for j in range(n):
            if i==j:
                mat[i][j] = 0
            elif input(f"Does edge exist between {i+1} and {j+1} y/n? ") == "y":
                mat[i][j] = int(input(f"Input {i+1}->{j+1} edge length : "))
            else:
                mat[i][j] = math.inf
    return mat

def PrintMat(mat):
    for i in mat:
        for j in i:
            print(j, end=" ")
        print("")
```

```python
def Warshall(mat):
    W = mat.copy()
    for k in range(len(W)):
        for i in range(len(W)):
            for j in range(len(W)):
                if W[i][j] > W[i][k]+W[k][j]:
                    W[i][j] = W[i][k]+W[k][j]
    return W
```

For the given graph we will, input the values for the weight matrix.


```python
n = int(input("Input number of vertices : "))
WgtMat = InpWgtMat(n)
print("\nWeight matrix of the graph : ")
PrintMat(WgtMat)
DisMat = Warshall(WgtMat)
print("\nShortest path between vertices is : ")
PrintMat(DisMat)
```
