# Insertion Sort

```python
from datetime import datetime as dt
import random
import time
```

```python
def arr_generate(n) -> list:
    return [random.randint(0, 2**15-1) for i in range(n)]
```

```python
def insertSort(arr):
    for i in range(1, len(arr)-2):
        v = arr[i]
        j = i-1
        while j>=0 and arr[j]>v:
            arr[j+1] = arr[j]
            j-=1
        arr[j+1] = v 
```

```python
%%timeit -n 50


arr = arr_generate(100)
insertSort(arr)
```

```python
start = time.time()
arr = arr_generate(100)
insertSort(arr)
print(f"Sorted Array : {arr}")
time_taken = time.time() - start
print(f"\nTime Taken : {time_taken*1000} ms")
```
