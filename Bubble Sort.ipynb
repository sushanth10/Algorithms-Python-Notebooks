# Bubble Sort

```python
from datetime import datetime as dt
import random
import time
```

```python
# Function that generates a list of n random integers

def arr_generate(n) -> list:
    return [random.randint(0, 2**15-1) for _ in range(n)]
```

```python
# Bubble Sort algorithm

def BubbleSort(arr) -> list:
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

```python
%%timeit -n 50

arr = arr_generate(100)
sorted_arr = BubbleSort(arr)
```

```python
start = time.time()
arr = arr_generate(100)
sorted_arr = BubbleSort(arr)
print(f"Sorted Array : {sorted_arr}")
time_taken = time.time() - start
print(f"\nTime Taken : {time_taken*1000} ms")
```
