# Quick Sort

```python
from datetime import datetime as dt
import time 
import random
```

```python
def arr_generate(n) -> list:
    return [random.randint(1, 2**15-1) for i in range(n)]
```

```python
def Quicksort(arr, l, r):
    if l<r:
        s = Partition(arr, l, r)
        Quicksort(arr, l, s)
        Quicksort(arr, s+1, r)
    
        
def Partition(arr, l, r):
    p = arr[l]
    i = l
    for j in range(l, r):
        if arr[j]<= p:
            arr[i], arr[j] = arr[j], arr[i]
            i+=1
    arr[i-1], arr[l] = arr[l], arr[i-1]
    return i-1
    
```

```python
%%timeit -n 50


arr = arr_generate(100)
Quicksort(arr, 0, len(arr))
```

```python
start = time.time()
arr = arr_generate(100)
Quicksort(arr, 0, len(arr))
print(f"Sorted Array : {arr}")
time_taken = time.time() - start
print(f"\nTime Taken : {time_taken*1000} ms")
```
