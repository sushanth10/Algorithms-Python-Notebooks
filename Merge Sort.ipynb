# Merge Sort

```python
from datetime import datetime as dt
import random
import time
```

```python
def arr_generate(n) -> list:
    return [random.randint(0, 2**15-1) for i in range(n)]
```

```python
def mergesort(arr):
    n = len(arr)
    if n>1:
        A = arr[:n//2]
        B = [i for i in arr if i not in A]
        mergesort(A)
        mergesort(B)
        Merge(A,B,arr)
        
def Merge(A, B, arr):
    i=j=k=0
    while(i<len(A) and j<len(B)):
        if A[i]<B[j]:
            arr[k] = A[i]
            i+=1
        else:
            arr[k] = B[j]
            j+=1
        k+=1
        
    if i==len(A):
        arr[k:] = B[j:]
    else:
        arr[k:] = A[i:]        
```

```python
%%timeit -n 50


arr = arr_generate(100)
mergesort(arr)
```

```python
start = time.time()
arr = arr_generate(100)
mergesort(arr)
print(f"Sorted Array : {arr}")
time_taken = time.time() - start
print(f"\nTime Taken : {time_taken*1000} ms")
```
