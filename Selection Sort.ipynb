# Selection Sort

```python
from datetime import datetime as dt
import random
import time
```

```python
# Function that generates a list of n random integers
def arr_generate(n) -> list:
    return [random.randint(0, 2**15-1) for i in range(n)]
```

```python
# Selection sort algorithm 

def SelectionSort(arr):
    n = len(arr)
    for i in range(0, n-1):
        min_pos = i
        for j in range(i+1, n):
            if arr[j]<arr[min_pos]:
                min_pos = j
        arr[i], arr[min_pos] = arr[min_pos], arr[i]
    return arr
```

```python
%%timeit -n 50

arr = arr_generate(100)
SelectionSort(arr)
```

```python
start = dt.now()
arr = arr_generate(100)
sorted_arr = SelectionSort(arr)
print("Sorted array : "+str(sorted_arr))
time_taken = dt.now() - start
print("\nTime Taken for one random execution : "+ str(time_taken.total_seconds()*1000) + " ms")
```
