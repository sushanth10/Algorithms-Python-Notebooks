# Euclid's Algorithm to Find GCD



```python
from datetime import datetime as dt
import random
import time
```

#### The following function generates 2 random integers between 2 and 500

```python
def two_rand_int() -> list:
    return [random.randint(2,500) for i in range(2)]
```

### The following is an iterative approach to find GCD using Euclid's Algorithm.

```python
# FUNCTION DEFINITION
def GCD(m, n) -> int:
    while(n != 0):
        r = m%n
        m = n
        n = r
    return m
```

```python
%%timeit -n 50

# timeit module is used to find the average time for execution
# -n argument is used to give the number of times the following lines to execute

num1, num2 = two_rand_int()
GCD(num1, num2)
```

```python
start = dt.now()
num1, num2 = two_rand_int()
print(f"GCD of {num1} and {num2} is {GCD(num1, num2)}")
time_taken = dt.now() - start
print("Time Taken : "+ str(time_taken.total_seconds()*1000)+"ms")
```
