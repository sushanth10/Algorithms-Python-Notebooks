# Warshall's Algorithm

This algorithm is used to find the transitive closure of a given graph using it's adjacency matrix.

```python
def InpAdjMatrix() -> list:
    try:
        n = int(input("Enter number of vertices : "))
        adjMat = [[0 for _ in range(n)] for _ in range(n)]
        print("Enter 1 if edge exists, otherwise 0")
        for i in range(n):
            for j in range(n):
                adjMat[i][j] = int(input(f"Between vertex {i+1} and {j+1}: "))
        return adjMat
    except ValueError:
        print("Invalid Value")
            
```

```python
# After each iteration of "k" loop we'll be tracking all the possible paths to and from each vertex and will add "1" wherever
# the paths are found between the two vertices.

def Warshall(mat) -> list:
    clos = mat.copy()
    for k in range(len(mat)):
        for i in range(len(mat)):
            for j in range(len(mat)):
                if mat[i][j]==1 or (mat[i][k]==mat[k][j]==1):
                    mat[i][j] = 1
    return clos
```
